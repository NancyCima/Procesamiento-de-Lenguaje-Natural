## **Web Scraping**

```python
URL = 'https://www.gutenberg.org/browse/scores/top1000.php#books-last1'
URL_ABS ='https://www.gutenberg.org'
```

```python
def get_summary(book_url: str) -> str:
    """
    Realiza una solicitud GET a la URL especificada, parsea el contenido HTML
    para obtener el resumen de un libro y lo devuelve como una cadena de texto.

    Parámetros:
    -----------
    book_url : str
        La URL de la página web de Gutenberg del libro del cual se extraerá el resumen del libro.
        En caso de no encontrar un resumen, usará el titulo del libro. En caso de no encontrar
        el titulo tampoco, retornará "Resumen no disponible".

    Retorna:
    --------
    summary : str
    """
    try:
        response = requests.get(book_url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error al acceder al resumen del libro en {book_url}: {e}")
        return "Resumen no disponible"

    response_parsed = BeautifulSoup(response.content, 'html.parser')
    summary_tag = response_parsed.find('th', string='Summary')

    if summary_tag:
        summary = summary_tag.find_next_sibling('td').get_text(strip=True)
        return summary
    else:
        title_tag = response_parsed.find('th', string='Title')
        if title_tag:
            title = title_tag.find_next_sibling('td').get_text(strip=True)
            return title
        else:
            return "Resumen no disponible"
```

```python
def get_books(url: str) -> List[Tuple[str, str]]:
    """
    Realiza una solicitud GET a la URL especificada, parsea el contenido HTML
    de la página para buscar todas las etiquetas <ol> (listas ordenadas) y
    extrae títulos de libros de la lista de libros que están dentro de etiquetas <li>.
    Además, extrae los links a los libros.
    Devuelve una lista de tuplas, con los títulos y links de todos los libros.

    Parámetros:
    -----------
    url : str
        La URL de la página web de Gutenberg desde la cual se extraerán las descripciones de libros.

    Retorna:
    --------
    books : List[Tuple[str, str]]
        Una lista de tuplas de cadenas de texto que contienen los títulos de los libros extraídos de la página web junto con los links a dichos libros.
    """
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error al acceder a la página web: {e}")
        return []

    response_parsed = BeautifulSoup(response.content, 'html.parser')

    ols = response_parsed.findAll('ol')
    books = []

    for ol in ols:
        book = [li.text for li in ol.findAll('li') if li.text != '']
        link = [a['href'] for a in ol.findAll('a')]
        books.extend(zip(book, link))

    return books
```

```python
books = get_books(URL)
books
```

```python
def create_csv(books: List[Tuple[str, str]], URL_ABS: str, SAVE_PATH: str) -> None:
    """
    Crea un archivo CSV a partir de una lista de libros y lo guarda en el directorio especificado.
    Para ello, extrae información clave de cada entrada de la lista, como el título principal, y,
    si estan dispinibles, el título secundario, el autor y el número de referencia.
    Además, se eliminan las filas duplicadas para asegurar la unicidad de los registros.
    Finalmente, se guarda el resultado en un archivo CSV en la ubicación que el usuario designe.

    Parámetros:
    -----------
    books : List[Tuple[str, str]]
        Una lista de cadenas de texto que contienen las descripciones de los libros.
    SAVE_PATH : str
        La ruta del directorio donde se guardará el archivo CSV.

    Retorna:
    --------
    None

    """

    data = []

    for book in books:
        book_info = book[0]
        book_link = book[1]

        # Utilizar una expresión regular para extraer la información
        match = re.match(r'^(.*?)(?:\s*:\s*(.*?))?\s+by\s+(.*?)\s+\((\d+)\)$', book_info)

        if match:
            titulo_principal = match.group(1).strip() if match.group(1) else None
            titulo_secundario = match.group(2).strip() if match.group(2) else None
            autor = match.group(3).strip() if match.group(3) else None
            numero_referencia = match.group(4).strip() if match.group(4) else None

            data.append({
                'Titulo Principal': titulo_principal,
                'Titulo Secundario': titulo_secundario,
                'Autor': autor,
                'Numero de Referencia': numero_referencia,
                'Descripción': get_summary(URL_ABS + book_link)
            })

    # Crear un DataFrame de pandas
    df = pd.DataFrame(data)

    # Eliminar filas duplicadas
    df.drop_duplicates(inplace=True)

    # Guardar el DataFrame en un archivo CSV
    df.to_csv(SAVE_PATH, index=False)
```

```python
create_csv(books, URL_ABS, 'data/dataset_books.csv')
```
